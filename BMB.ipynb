```python
import os
import pandas as pd
import camelot
from PyPDF2 import PdfReader

# === CONFIGURATION ===
# Define the folder where your PDFs are stored
pdf_folder = r"C:\Users\Dell\Downloads\Malaysia Islamic Banks\ICM Banks ASS\BMB"

# Define the Excel file to save extracted tables
excel_file = os.path.join(pdf_folder, "BMB_Financial_Data_2010_2024.xlsx")

# Define the file paths for each year
pdf_files = {
    "2010": os.path.join(pdf_folder, "BMB_2010.pdf"),
    "2012": os.path.join(pdf_folder, "BMB_2012.pdf"),
    "2014": os.path.join(pdf_folder, "BMB_2014.pdf"),
    "2015": os.path.join(pdf_folder, "BMB_2015.pdf"),  # FIXED file path issue
    "2016": os.path.join(pdf_folder, "BMB_2016.pdf"),
    "2017": os.path.join(pdf_folder, "BMB_2017.pdf"),  # FIXED: was using 2018
    "2018": os.path.join(pdf_folder, "BMB_2018.pdf"),
    "2019": os.path.join(pdf_folder, "BMB_2019.pdf"),
    "2020": os.path.join(pdf_folder, "BMB_2020.pdf"),  # FIXED: was using 2021
    "2021": os.path.join(pdf_folder, "BMB_2021.pdf"),
    "2022": os.path.join(pdf_folder, "BMB_2022.pdf"),  # FIXED: was using 2023
    "2023": os.path.join(pdf_folder, "BMB_2023.pdf"),
    "2024": os.path.join(pdf_folder, "BMB_2024.pdf"),
}

# Define the page numbers to extract for each year (customize as needed)
page_numbers = {
    "2010": "42",
    "2012": "58",
    "2014": "72",
    "2015": "72",
    "2016": "71",
    "2017": "81",
    "2018": "80",
    "2019": "90",
    "2020": "84",
    "2021": "81",
    "2022": "83",
    "2023": "82",
    "2024": "20",
}

# === FUNCTION TO CHECK PDF PAGE COUNT ===
def get_pdf_page_count(pdf_path):
    """Returns the total number of pages in a PDF file."""
    try:
        reader = PdfReader(pdf_path)
        return len(reader.pages)
    except Exception as e:
        print(f"❌ Error reading {pdf_path}: {e}")
        return 0

# === FUNCTION TO EXTRACT TABLES FROM PDF ===
def extract_tables(pdf_path, page):
    """Extracts tables from a given PDF page using Camelot."""
    try:
        # Check if the PDF exists
        if not os.path.exists(pdf_path):
            print(f"❌ PDF not found: {pdf_path}")
            return []

        # Get total pages
        total_pages = get_pdf_page_count(pdf_path)

        # Check if requested page is valid
        if int(page) > total_pages:
            print(f"⚠️ Warning: Page {page} does not exist in {pdf_path}. Skipping...")
            return []

        # Extract tables
        tables = camelot.read_pdf(pdf_path, pages=page, flavor="stream")

        if tables.n == 0:
            print(f"⚠️ No tables found on page {page} in {pdf_path}.")
        else:
            print(f"✅ Extracted {tables.n} tables from page {page} in {pdf_path}")

        return tables

    except Exception as e:
        print(f"❌ Error extracting tables from {pdf_path}: {e}")
        return []

# === MAIN PROCESS TO EXTRACT AND SAVE DATA ===
writer = pd.ExcelWriter(excel_file, engine="openpyxl")

for year, pdf_path in pdf_files.items():
    if year in page_numbers:
        tables = extract_tables(pdf_path, page_numbers[year])

        for i, table in enumerate(tables):
            df = table.df  # Convert to DataFrame
            sheet_name = f"{year}_Table{i}"

            # Ensure sheet name is within Excel's 31-character limit
            if len(sheet_name) > 31:
                sheet_name = f"{year}T{i}"

            df.to_excel(writer, sheet_name=sheet_name, index=False)
            print(f"📄 Saved {year} Table {i} to Excel.")

# Save the Excel file
writer.close()
print(f"\n✅ All tables successfully saved to: {excel_file}")
```
