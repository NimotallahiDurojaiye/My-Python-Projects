```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import time

# Step 1: Setup Chrome options (optional)
options = Options()
# options.add_argument("--headless")  # Uncomment to run browser in background

# Step 2: Initialize webdriver (Selenium Manager handles driver)
driver = webdriver.Chrome(options=options)

# Step 3: Open LinkedIn login page
driver.get("https://www.linkedin.com/login")

# Step 4: Pause to let you manually login (for security)
input("Log into LinkedIn manually, then hit Enter here to continue...")

# Step 5: Perform keyword content search
keywords = "Islamic Finance AI upskilling automation Shariah compliance training"
search_query = "%20".join(keywords.split())
url = f"https://www.linkedin.com/search/results/content/?keywords={search_query}&origin=GLOBAL_SEARCH_HEADER"
driver.get(url)
time.sleep(5)

# Step 6: Scroll page to load more content
for _ in range(3):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)

# Step 7: Parse page content with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')
post_divs = soup.find_all('div', class_='feed-shared-update-v2')  # Adjust if LinkedIn UI changes

# Step 8: Filter posts by your theoretical keywords
variables_keywords = [
    'ai skill', 'upskilling', 'training', 'automation', 'task complexity',
    'shariah compliance', 'technical skill', 'transversal skill', 'organizational support',
    'community', 'talent development', 'policy update'
]

def contains_keywords(text):
    text = text.lower()
    return any(kw in text for kw in variables_keywords)

posts = []
for div in post_divs:
    post_text = div.get_text(separator=' ', strip=True)
    if contains_keywords(post_text):
        posts.append(post_text)

# Step 9: Save extracted posts to CSV file
df = pd.DataFrame({'Post Content': posts})
df.to_csv("linkedin_ifi_ai_posts.csv", index=False)

print(f"Extracted {len(posts)} posts saved to linkedin_ifi_ai_posts.csv")

# Step 10: Close webdriver
driver.quit()
```

```python
import random
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import time

# Setup Chrome options
options = Options()
# Uncomment below for headless operation
# options.add_argument("--headless")
options.add_argument("--start-maximized")

# Initialize WebDriver (Selenium Manager handles driver)
driver = webdriver.Chrome(options=options)

# Step 1: Open LinkedIn login
driver.get("https://www.linkedin.com/login")
input("Please login manually, then press Enter here to continue...")

# Step 2: Define search keywords slightly broader initially
keywords = [
    'ai skill', 'upskilling', 'training', 'automation', 'task complexity',
    'shariah compliance', 'technical skill', 'transversal skill', 'organizational support',
    'community', 'talent development', 'policy update', 'customer engagement',
    'risk analytics', 'kyc', 'rogue ai', 'ethics', 'fraud detection', 'governance'
]
search_query = "%20".join(keywords)
url = f"https://www.linkedin.com/search/results/content/?keywords={search_query}&origin=GLOBAL_SEARCH_HEADER"
driver.get(url)
time.sleep(5)

# Step 3: Scroll multiple times with random delays to allow content loading
scroll_pause_time = 3
for _ in range(5):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(scroll_pause_time + random.uniform(0, 2))

# Step 4: Parse page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# LinkedIn post container class may change; try multiple selectors
post_divs = soup.find_all('div', class_='feed-shared-update-v2')
if not post_divs:
    post_divs = soup.find_all('div', class_='occludable-update')

print(f"Found {len(post_divs)} posts on page")

# Step 5: Filter posts for presence of any keyword
def contains_keywords(text, keyword_list):
    text = text.lower()
    return any(kw in text for kw in keyword_list)

posts = []
for div in post_divs:
    text = div.get_text(separator=' ', strip=True)
    if contains_keywords(text, keywords):
        posts.append(text)

print(f"Filtered to {len(posts)} relevant posts")

# Step 6: Save to CSV
df = pd.DataFrame({'Post Content': posts})
filename = "linkedin_ifi_ai_posts.csv"
df.to_csv(filename, index=False)
print(f"Saved posts to {filename}")

# Step 7: Cleanup
driver.quit()
```
