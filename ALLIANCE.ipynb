```python
import pandas as pd
import glob
import os

# Find all CSV files in the current directory
csv_files = glob.glob("*.csv")

# Create an Excel writer object to save multiple sheets
with pd.ExcelWriter("ALLIANCE_sheets.xlsx", engine="openpyxl") as writer:
    for file in csv_files:
        df = pd.read_csv(file)
        # Use the CSV filename (without extension) as the sheet name
        sheet_name = os.path.splitext(os.path.basename(file))[0]
        # Save the DataFrame to a sheet in the workbook
        df.to_excel(writer, sheet_name=sheet_name, index=False)

print("Merged Excel workbook saved as 'ALLIANCE_sheets.xlsx'.")
```

```python
# List of years
years = ["2017A", "2018A", "2019A", "2020A", "2021A", "2022A", "2023A", "2024A"]

# Save each dataframe to CSV
for year in years:
    table = globals().get(f"ALLIANCE{year}")  # Get the variable dynamically
    if table and isinstance(table, list):  # Ensure it's a list (from Camelot)
        table[0].df.to_csv(f"{year}.csv", index=False)  # Save first extracted table only
        print(f"Saved {year}.csv")
```

```python
output_file = "ALLIANC15_MERGED.xlsx"

with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
    for year in years:
        file_name = f"{year}.csv"
        try:
            df = pd.read_csv(file_name)  # Read CSV
            df.to_excel(writer, sheet_name=year, index=False)  # Save to respective sheet
            print(f"Added {year} data to {output_file}")
        except FileNotFoundError:
            print(f"⚠️ Warning: {file_name} not found, skipping...")

print(f"✅ All data saved in {output_file}")
```
