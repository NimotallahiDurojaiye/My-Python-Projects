```python
!pip install pyvis
from pyvis.network import Network
import networkx as nx

# Initialize graph
G = nx.Graph()
nt = Network(height="1000px", width="100%", notebook=True, cdn_resources='remote')
nt.barnes_hut()

# Root node
root_node = "Automation Pipelines"
G.add_node(root_node, size=30, color="#FF6B6B")

# Pipeline tiers with pricing and links
tiers = {
    "Basic (Free/Low Cost)": {
        "color": "#4ECDC4",
        "tools": {
            "Recording/Transcription": [
                {"name": "Otter.ai (Free)", "url": "https://otter.ai", "cost": "$0–$16.99/month"},
                {"name": "Fireflies.ai (Free)", "url": "https://fireflies.ai", "cost": "$0–$18/month"}
            ],
            "Summarization": [
                {"name": "ChatGPT (Free)", "url": "https://chat.openai.com", "cost": "Free"},
                {"name": "Notion AI ($10)", "url": "https://notion.so", "cost": "$10/month"}
            ],
            "Analysis": [
                {"name": "QCAmap (Free)", "url": "https://www.qcamap.org", "cost": "Free"},
                {"name": "Thematic (Trial)", "url": "https://getthematic.com", "cost": "$200+/month"}
            ]
        }
    },
    "Mid-Range": {
        "color": "#45B7D1",
        "tools": {
            "Recording/Transcription": [
                {"name": "Rev.ai ($0.25/min)", "url": "https://rev.ai", "cost": "$0.25/minute"},
                {"name": "Sonix ($10/hr)", "url": "https://sonix.ai", "cost": "$10/hour"}
            ],
            "Summarization": [
                {"name": "Fireflies Pro ($18)", "url": "https://fireflies.ai", "cost": "$18/month"},
                {"name": "MS Copilot ($30)", "url": "https://microsoft.com", "cost": "$30/month"}
            ],
            "Analysis": [
                {"name": "Dovetail ($15–$75)", "url": "https://dovetail.com", "cost": "$15–$75/month"},
                {"name": "MAXQDA ($36/mo)", "url": "https://maxqda.com", "cost": "$36/month"}
            ]
        }
    },
    "Advanced/Enterprise": {
        "color": "#96CEB4",
        "tools": {
            "Recording/Transcription": [
                {"name": "Trint ($48)", "url": "https://trint.com", "cost": "$48/month"},
                {"name": "Verbit ($1.50–$3/min)", "url": "https://verbit.ai", "cost": "$1.50–$3/minute"}
            ],
            "Summarization": [
                {"name": "Gong ($125+)", "url": "https://gong.io", "cost": "$1,500+/year"},
                {"name": "MS 365 Copilot ($30/user)", "url": "https://microsoft.com", "cost": "$30/user/month"}
            ],
            "Analysis": [
                {"name": "NVivo ($58+/mo)", "url": "https://qsrinternational.com", "cost": "$58+/month"},
                {"name": "Dedoose ($12.95+)", "url": "https://dedoose.com", "cost": "$12.95+/month"}
            ]
        }
    },
    "Custom AI (API-Based)": {
        "color": "#FFEEAD",
        "tools": {
            "Recording/Transcription": [
                {"name": "OpenAI Whisper (Free)", "url": "https://openai.com", "cost": "Free"},
                {"name": "AWS Transcribe ($0.024/min)", "url": "https://aws.amazon.com", "cost": "$0.024/minute"}
            ],
            "Summarization": [
                {"name": "GPT-4 API ($0.06/1K tokens)", "url": "https://openai.com", "cost": "$0.06/1K tokens"}
            ],
            "Analysis": [
                {"name": "Custom NLP ($50–$500)", "url": "#", "cost": "$50–$500/month"}
            ]
        }
    }
}

# Build the graph
G.add_node(root_node, size=30, color="#FF6B6B")
for tier, data in tiers.items():
    # Add tier node
    G.add_node(tier, color=data["color"], size=25)
    G.add_edge(root_node, tier)
    
    # categories and tools
    for category, tools in data["tools"].items():
        G.add_node(category, color="#D3D3D3", size=20)
        G.add_edge(tier, category)
        
        for tool in tools:
            # clickable link
            tool_name = tool["name"]
            tool_url = tool["url"]
            G.add_node(
                tool_name,
                title=f'<a href="{tool_url}" target="_blank">{tool_name}</a>',
                color="white",
                size=15
            )
            G.add_edge(category, tool_name)
            
            # Cost node
            cost = tool["cost"]
            G.add_node(cost, color="#F9F9F9", size=12, shape="box")
            G.add_edge(tool_name, cost)


nt.set_options("""
{
  "physics": {
    "forceAtlas2Based": {
      "springLength": 150,
      "springConstant": 0.01,
      "damping": 0.5
    },
    "minVelocity": 0.75,
    "solver": "forceAtlas2Based"
  }
}
""")


nt.from_nx(G)

# Save and open
nt.show("automation_pipelines_enhanced.html")
```
