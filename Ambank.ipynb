```python
import pandas as pd
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows

# Define the output path
output_path = r"C:\Users\Dell\Downloads\Malaysia Islamic Banks\ICM Banks ASS\AMBANK ISLAMIC\Ambank_Financial_Data_2015_2024.xlsx"
# Create a Pandas Excel writer using openpyxl as the engine
excel_file = 'Ambank_Financial_Data_2010_2023.xlsx'
writer = pd.ExcelWriter(excel_file, engine='openpyxl')

# Dictionary to store all tables
all_tables = {
    "2015": Ambank2015,
    "2016": Ambank2016,
    "2017": Ambank2017,
    "2018": Ambank2018,
    "2019": Ambank2019,
    "2020": Ambank2020,
    "2021": Ambank2021,
    "2022": Ambank2022,
    "2023": Ambank2023,
    "2024": Ambank2024
}

# Write each table to a different worksheet
for year, tables in all_tables.items():
    for i, table in enumerate(tables):
        sheet_name = f'{year}_Table{i}'
        # Ensure sheet name is valid (31 chars max for Excel)
        if len(sheet_name) > 31:
            sheet_name = f'{year}T{i}'
            
        # Convert table to dataframe and write to Excel
        df = table.df
        df.to_excel(writer, sheet_name=sheet_name, index=False)
        print(f"Added {year} Table {i} to Excel file")

# Save the Excel file
writer.close()
print(f"All tables successfully saved to {excel_file}")
```
