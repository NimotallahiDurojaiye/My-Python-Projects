```python
import pandas as pd
import numpy as np
from linearmodels.panel import PanelOLS
import statsmodels.api as sm

# Simulate minimal data
data = pd.DataFrame({
    'entity': [1,1,2,2],
    'time': [1,2,1,2],
    'y': [1.1, 2.3, 1.9, 3.1],
    'x': [0.5, 1.2, 0.8, 1.5]
}).set_index(['entity', 'time'])

mod = PanelOLS(data['y'], data[['x']], entity_effects=True)
res = mod.fit()
print(res)
```
