```python
import yfinance as yf

# Define the list of ticker symbols
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'HLAL', 'BABA', 'FFB', 
           'PMETAL', 'TENAGA', 'TOPGLOV', 'PYPL', 'NVDA', 'META', 'EFA',
           'XOM', 'MMM', 'ADBE', 'AMD', 'GOOG', 'ASML', 'BURSA', 'CSCO',
           'LLY', 'BAMUDA', 'IHH', 'INTU', 'ISDW', 'NVO', 'PCHEM', 'PG',
           'QL', 'RANHILL', 'CRM', 'SIME', 'TSM', 'T;', 'TXN', 'V', 'VISA']

# Download data for multiple tickers
data = yf.download(tickers, period='6mo')['Adj Close']

# Display the data
print(data.head())
```

```python
# Calculate daily returns
returns = data.pct_change().dropna()

```

```python
# Calculate expected returns and risk (standard deviation)
expected_returns = returns.mean() * 126  # Annualize returns
risk = returns.std() * np.sqrt(12)  # Annualize risk

# Display results
print("Expected Returns:\n", expected_returns)
print("Risk:\n", risk)
```
